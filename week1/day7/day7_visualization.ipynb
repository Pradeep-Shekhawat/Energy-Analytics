## 2. Importing Libraries & Data

In this section, I will:

- Import `pandas` and `matplotlib.pyplot`  
- Load my `solar_irr.csv` from Day 6 with `pd.read_csv(..., parse_dates=["date"])`  
- Set `date` as the DataFrame index  

*Notes:*  
- Use `plt.figure(figsize=(10,4))` for clear visuals  
- Always label axes, titles, and add a grid

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../day6/solar_irr.csv", parse_dates=["date"])
df.set_index("date", inplace=True)
```

---

## 3. Basic Line Plot

In this section, I will:

- Plot daily `irradiance_kwh` vs. date  
- Title the chart “Daily Solar Irradiance (kWh/m²)”  
- Label the x-axis “Date” and y-axis “Irradiance (kWh/m²)”  

```python
plt.figure(figsize=(10, 4))
plt.plot(df.index, df["irradiance_kwh"])
plt.title("Daily Solar Irradiance (kWh/m²)")
plt.xlabel("Date")
plt.ylabel("Irradiance (kWh/m²)")
plt.grid(True)
plt.show()
```

---

## 4. Rolling Statistics (7-Day Moving Average)

In this section, I will:

- Compute a 7-day rolling mean of `irradiance_kwh`  
- Overlay the raw series and its moving average  
- Add a legend distinguishing “Daily” and “7-Day MA”  

```python
rolling = df["irradiance_kwh"].rolling(window=7).mean()
plt.figure(figsize=(10, 4))
plt.plot(df.index, df["irradiance_kwh"], label="Daily", alpha=0.6)
plt.plot(df.index, rolling, label="7-Day MA", linewidth=2)
plt.title("Solar Irradiance with 7-Day Moving Average")
plt.xlabel("Date")
plt.ylabel("Irradiance (kWh/m²)")
plt.legend()
plt.show()
```

---

## 5. Histogram & Distribution

In this section, I will:

- Create a histogram of daily `irradiance_kwh` values with 20 bins  
- Title the plot “Distribution of Daily Irradiance”  
- Label axes and display frequency  

```python
plt.figure(figsize=(8, 4))
plt.hist(df["irradiance_kwh"], bins=20)
plt.title("Distribution of Daily Irradiance")
plt.xlabel("Irradiance (kWh/m²)")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

---

## 6. Save All Plots to Files

In this cell, I will:

1. Generate each of the three plots above.  
2. After each plot, call `plt.savefig("<filename>.png")` before closing.  
3. Verify that `line_plot.png`, `rolling_mean.png`, `histogram.png` appear in my folder.  

```python
# Line Plot:
plt.figure(figsize=(10, 4))
plt.plot(df.index, df["irradiance_kwh"])
plt.title("Daily Solar Irradiance (kWh/m²)")
plt.xlabel("Date")
plt.ylabel("Irradiance (kWh/m²)")
plt.grid(True)
plt.savefig("line_plot.png")
plt.close()

# Rolling Mean:
rolling = df["irradiance_kwh"].rolling(window=7).mean()
plt.figure(figsize=(10, 4))
plt.plot(df.index, df["irradiance_kwh"], label="Daily", alpha=0.6)
plt.plot(df.index, rolling, label="7-Day MA", linewidth=2)
plt.title("Solar Irradiance with 7-Day Moving Average")
plt.xlabel("Date")
plt.ylabel("Irradiance (kWh/m²)")
plt.legend()
plt.savefig("rolling_mean.png")
plt.close()

# Histogram:
plt.figure(figsize=(8, 4))
plt.hist(df["irradiance_kwh"], bins=20)
plt.title("Distribution of Daily Irradiance")
plt.xlabel("Irradiance (kWh/m²)")
plt.ylabel("Frequency")
plt.grid(True)
plt.savefig("histogram.png")
plt.close()
```
